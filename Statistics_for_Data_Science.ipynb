## Hypothesis Testing

### Definisi

* Hypothesis Testing
  Pendekatan statistik dengan melakukan tes terhadap asumsi yang berkaitan dengan populasi parameter

* Null Hypothesis (H0)
  Menyatakan bahwa tidak ada statistical difference terhadap suatu karakteristik tertentu dari suatu populasi

* Alternative Hypothesis
  Menyatakan terdapat statistical difference terhadap suatu karakteristik tertentu dari suatu populasi

* Significance Level
  Simbol = alpha

* Confidence Level
  Simbol = 1 - alpha
  Confidence Level menunjukkan peluang dimana jika studi dilakukan secara berulang ulang, hasil yang didapatkan akan sama

* Confidence Interval
  Spread yang menunjukkan range dari nilai means dari suatu eksperimen, menunjukkan bahwa hasil eksperimen berulang akan sama dengan Confidence Interval tersebut.

### Type Error

* **Type 1 Error(alpha)**
  Probability dari rejecting Null Hypothesis(H0), walaupun ia benar.

  Nama lain = False Positive
  Berkaitan dengan Significance Level

* **Type 2 Error(Beta)**

  Probability dari tidak mereject Null Hypothesis, walau dia tidak benar

  Nama lain = False Negative
  Berkaitan dengan Statistical Power

### Jenis Hypothesis Testing

**Parametric Test**
  Jenis dimana memiliki kondisi prior knowledge terhadap distribusi dari populasinya

  Parameter berupa :  
  Mean
  Variance

  Jenis Parametric Test :
  T-Test
  Z-Test
  ANOVA

**Non Parametric Test**

Diimplementasikan tanpa adanya asumsi terhadap distribusi populasi

Parameter berupa : Median

Jenis Test : Kruskal Wallis, Mann Whitney U, Chi-Square (Data Kategorik)
